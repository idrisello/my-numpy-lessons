### Binary funksiyalar (ikkita argument qabul qiluvchi funkisya)
### `add`, `multiply`, `maximum` funksiyalarini ko'rib chiqamiz


```python
# 2 ta massivni yaratib olamiz
arr2 = np.random.randn(6)
arr3 = np.random.randn(6)
arr2
```

```python
arr3
```

```python
np.add(arr2, arr3) # arr2 va arr3 ning elementlarini mos ravishda qo'shib beradi
```

```python
arr2 + arr3 # arr2 va arr3 ning elementlarini mos ravishda qo'shib beradi
```

```python
np.multiply(arr2, arr3) # arr2 va arr3 ning mos elementlarini ko'paytiradi
```

```python
np.maximum(arr2, arr3) # arr2 va arr3 larning elementlarin taqqoslab, kattasining elementini qaytaradi
```

```python
# Massivlar
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5]) # xarr
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5]) # yarr
```

```python
# Shart
cond = np.array([True, False, True, True, False])
```

```python
# where
result = np.where(cond, xarr, yarr) # cond==True : xarr elementi, cond==False : yarr elementi
result
```

```python
arr001 = np.random.randn(4, 4)  # (4, 4) taxminiy massiv
arr001
```

```python
# where
results = np.where(arr001<0, -2, 2) # arr ning manfiy elementlarini -2 ga musbat elementlarin esa 2 ga almashitirish
results
```

# **SORTING**

```python
arr19
```

```python
# 1 usul
sorted_arr = np.sort(arr19) # arr massivini tartiblangan ko'rinishini sorted_arr joylaydi
sorted_arr
```

```python
#  2 usul
arr19.sort() # arr ning tartiblangan ko'rinishi
arr19
```

```python
# teskari tartiblash
rev_arr = -np.sort(-arr19) # teskari tartibdagi arr
rev_arr
```

```python
# axis = 0 --> Qator
row_srt = np.sort(arr19, axis=0) # arr19 ning xar bir qator bo'yicha tartiblangan ko'rinishi
row_srt
```

```python
# axis =1 --> Ustun
col_srt =np.sort(arr19, axis=1) # arr2d ning xar bir ustun bo'yicha tartiblangan ko'rinishi
col_srt
```
