```python
#When a vector is multiplied by a scalar, each element of the vector is multiplied with the scalar
v1=np.array([1,2,3,4])
v1*2
```

```python
#Another common operation is dot product. The dot product of two vectors is the sum of the products of elements with regards to position.
# The first element of one vector is multiplied by the first element of the second vector and so on.
# The sum of these products is the dot product which can be done with np.dot() function.
vector1=np.array([1,2,3])
vector2=np.array([4,5,6])
np.dot(vector1,vector2) #1vektor va 2-vektor elementlari mos ravisdha kopaytrlmoqda va kopaygan jovobla qoshilgan
```

```python
print('vektor1 va vektor2ni  yigindisi{}'.format(vector1+vector2))
```

```python
print('vektor1 va vektor2ni  ayirmasi{}'.format(vector1-vector2))
```

Matrix multiplication of two matrices is based on dot products of rows of one matrix with the columns of other matrix. Each element of the product matrix is a dot product of a row in first matrix and a column in the second matrix. Consider matrices A1 and A2 below. The first row of A1 is [2,2] and the first column of A2 is [1,4]. Dot product of these two vectors is (2x1)+(2x4)=10 which is the element of the product matrix at position [0,0]. The dot product of third row of A1 and third column of A2 gives us the element at position [2,2] which is 12.

Since the rows of the first matrix are multiplied by the columns of the second matrix, the number of rows of first matrix must be equal to the number of columns of the second matrix. If we multiply mxn matrix (m rows, n columns) with nxk matrix, the shape of the product matrix is mxk:
